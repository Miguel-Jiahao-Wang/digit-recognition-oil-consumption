# Augmenting original digital images  
This notebook is used to create augmented versions of the original "digital gas pump display" pictures.

```
from imgaug import augmenters as iaa
from matplotlib.pyplot import imshow
from PIL import Image

import matplotlib.pyplot as plt
import numpy as np
import glob
import cv2

%matplotlib inline
```

```
# defining augmentation sequence

seq = iaa.Sequential([
    iaa.CropAndPad(percent=(-0.10, 0.10)),
    iaa.AverageBlur(k=(2, 25)),
    iaa.AdditiveGaussianNoise(scale=(0, 0.05*255)),
    iaa.Sometimes(0.25, iaa.CoarseDropout((0.0, 0.05), size_percent=(0.2, 0.25))),
    iaa.Affine(rotate=(-15,15),
               shear=(-15,15)),
    iaa.Add((-10,10), per_channel=0.5)
], random_order=True)
```

```
# defining method to create 'n_augmentations' number of augmentations of each image

def augment_data(input_dir, output_dir, augmentation_seq, n_augmentations):
  files = glob.glob("{}*.jpg".format(input_dir))
  
  for file in files:
    fname = file.split("/")[-1]
    img = cv2.imread(file)
    
    for i in range(n_augmentations):
      aug = seq(images=[img])[0]
      new_fname = fname.replace(".jpg", "_{}.jpg".format(i))
      
      cv2.imwrite(output_dir + new_fname, aug)
```
