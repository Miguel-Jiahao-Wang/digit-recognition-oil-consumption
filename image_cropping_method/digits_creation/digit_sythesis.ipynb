# Digit synthesis with different fonts

This notebook is used to create new digit pictures by calling different fonts.

Reference: https://tanmayshah2015.wordpress.com/2015/12/01/synthetic-font-dataset-generation/

```
from PIL import Image, ImageDraw, ImageFont
import string
import ntpath
import numpy as np
import os
import glob
```

```
fontSize = 100
imgSize = (60,100)
position = (30,50)
 
#All images will be stored in 'Synthetic_dataset' directory under current directory
dataset_path = os.path.join (os.getcwd(), 'Synthetic_dataset')
if not os.path.exists(dataset_path):
   os.makedirs(dataset_path)
 
fhandle = fonts_list
digits = range(0,10)
digits_list=[]
for d in digits:
    digits_list.append(str(d))
all_char_list = digits_list # The characters to be sythesized
```

```
total_fonts = len(fonts_list)

#Collect all the fonts in a local machine
all_fonts = glob.glob("/System/Library/Fonts/*ttf")
all_fonts += glob.glob("/System/Library/Fonts/*ttc")
all_fonts += glob.glob("/System/Library/Fonts/*dfont")
all_fonts += glob.glob("/System/Library/Fonts/*otf")
all_fonts += glob.glob("/Users/Laiochen/Desktop/Projects/Digit_recognition/Font/FontFIle/*ttf")
all_fonts += glob.glob("/Users/Laiochen/Desktop/Projects/Digit_recognition/Font/FontFIle/*ttc")
all_fonts += glob.glob("/Users/Laiochen/Desktop/Projects/Digit_recognition/Font/FontFIle/*dfont")
all_fonts += glob.glob("/Users/Laiochen/Desktop/Projects/Digit_recognition/Font/FontFIle/*otf")

f_flag = np.zeros(total_fonts)
```

```
for sys_font in all_fonts:
    #print "Checking "+p
    font_file = ntpath.basename(sys_font)
    font_file = font_file.rsplit('.')
    font_file = font_file[0]
    f_idx = 0
    for font in fonts_list:
        f_lower = font.lower()
        s_lower = sys_font.lower()
        #Check desired font
        try:
            if f_lower in s_lower:
                path = sys_font
                font = ImageFont.truetype(path, fontSize) # Get the font
                f_flag[f_idx] = 1
                for ch in all_char_list:
                    image = Image.new("RGB", imgSize, (255,255,255))
                    draw = ImageDraw.Draw(image)
                    pos_x = 0
                    pos_y = 0
                    pos_idx=0
                    position = (pos_x,pos_y) # Position of the digits
                    draw.text(position, ch, (0,0,0), font=font) # Draw on the pic
                    file_name = font_file + '_' + ch + '.jpg' 
                    file_name = os.path.join(dataset_path,file_name) # Filename
                    image.save(file_name)
        except OSError:
            pass
    f_idx = f_idx + 1
```
